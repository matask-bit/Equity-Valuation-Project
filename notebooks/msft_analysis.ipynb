# MSFT Valuation Analysis

This notebook reproduces key tables and charts generated by the valuation script for Microsoft (MSFT).

Run `scripts/msft_valuation.py` first to generate the data and charts in the `outputs` directory.


```python
import os

import pandas as pd
from IPython.display import display, Image

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath("")))
OUTPUT_DIR = os.path.join(BASE_DIR, "outputs")

fin_path = os.path.join(OUTPUT_DIR, "msft_financial_summary.csv")
dcf_path = os.path.join(OUTPUT_DIR, "dcf_scenarios.csv")

fin_df = pd.read_csv(fin_path)
dcf_df = pd.read_csv(dcf_path)

print("Financial summary (head):")
display(fin_df.head())

print("\nDCF scenarios:")
display(dcf_df)
```

```python
# Show charts inline
rev_chart = os.path.join(OUTPUT_DIR, "revenue_chart.png")
fcf_chart = os.path.join(OUTPUT_DIR, "fcf_chart.png")
margins_chart = os.path.join(OUTPUT_DIR, "margins_chart.png")

print("Revenue chart:")
display(Image(filename=rev_chart))

print("Free cash flow chart:")
display(Image(filename=fcf_chart))

print("Margins chart:")
display(Image(filename=margins_chart))
```
